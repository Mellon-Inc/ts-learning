## <font color="SeaGreen">1-a 時系列データの定義と特徴</font>

### **定義**
時系列データとは、時間の経過とともに観測または記録されたデータのことを指します。これらのデータは一定の時間間隔（例：毎分、毎時間、毎日など）で収集され、観測された順序に意味があり、関係性を持っています。

#### <u>時系列データの例</u>
- 気象データ：気温、湿度、降水量（例：1時間ごと、1日ごと）
- 経済データ：株価の変動（例：分単位、日単位の価格変化）
- 健康データ：心拍数、血圧（例：秒単位、分単位）

#### <u>時系列データではない例</u>
- 顧客データ：顧客情報を記録したデータベース（時間軸に依存しない）
- アンケートデータ：満足度調査（回答順序が重要でない）
- 画像データ：写真や静止画像（ただし、動画データは時系列データに該当）

### **ポイント**
時系列データの判断基準は「<font color="red">時間の順序が意味を持つデータかどうか</font>」です。

---

### **特徴**
時系列データには、以下のような主な特徴があります：

#### <u>トレンド（Trend）</u>
データが長期的に上昇または下降する傾向を示します。  
例：
- 気温：世界の平均気温が年々上昇
- 消費量：電気自動車の普及に伴うガソリン消費量の減少

#### <u>季節性（Seasonality）</u>
一定の周期で繰り返されるパターンを指します。これにより、将来の周期的変動を予測しやすくなります。  
例：
- 小売業の売上：年末やクリスマス時期の売上急増
- 気温：夏の暑さ、冬の寒さ

#### <u>ランダム性（Noise）</u>
予測不可能な変動部分で、外的要因や偶然によるものです。  
例：
- 為替レート：政策変更や自然災害による突発的な変動
- センサー計測値：環境条件や計測機器の誤差

---

### **まとめ**
時系列データは「時間の経過に応じた観測データ」であり、トレンド・季節性・ランダム性などの特徴を理解することで、重要な情報を発見することができ、データ分析や予測の精度を高めることができます。


## <font color="SeaGreen">1-b 時系列データの用途と重要性</font>
時系列データは、時間の流れに沿って記録されたデータで、私たちの日常生活やビジネス、経済、科学、工学など幅広い分野で非常に重要な役割を果たしています。
時系列データにより、過去を理解し、未来を予測し、異常を発見することにより、さまざまな分野で効率的かつ効果的な運営や意思決定が可能になります。データに基づく判断が求められる現代社会において、時系列データの活用は欠かせないものとなっています。

#### <u>予測</u>
過去のデータに基づいて未来の状況を予測するのに使われます。  
例：
- 天気予報：過去の気象データ（気温、降水量、風速など）を分析して、明日や来週の天気を予測
- 売上予測：過去の売上データをもとに次の月や季節の売上を予測し、商品在庫を調整

#### <u>季節性の特定</u>
データの中に一定の周期で繰り返されるパターン（季節性）を見つけ出すことにより、周期的な変動を理解し、将来の計画や予測に役立てることができます。  
例：
- エネルギー消費量の分析：電力会社が発電計画を立て、ピーク時の供給不足を防止
- 観光業の需要予測：航空会社が需要の多い時期にフライトを増便

#### <u>トレンド分析</u>
長期的な変動を理解することで、未来に備えた効果的な対策や戦略を立てることが可能になり、潜在的なリスクを事前に把握したり、新しい市場や機会をいち早くつかむことができます。  
例：
- 感染症のトレンド：インフルエンザや新型ウイルスの感染者数データを分析し、感染の広がりやピークを予測
- ハッシュタグやトピックのトレンド：ソーシャルメディア上のハッシュタグや検索ワードの使用頻度を分析し、流行テーマに基づいて広告キャンペーンを展開

#### <u>異常検知</u>
通常のパターンから外れるデータ点（異常値）や異常な振る舞いを見つけることにより、システムの問題やリスクを早期に発見し、対策を講じることができます。  
例：
- 機械設備の異常検知：工場の機械の振動データや温度データをリアルタイムで監視し、通常のパターンから外れた異常な動きを検知
- システム障害の検知：サーバーログやアプリケーションの応答時間データを監視し、通常と異なる挙動を検知


## <font color="SeaGreen">1-c 時系列データとその他のデータ形式との違い</font>
時系列データは、時間という特定の次元を中心に分析されるため、他のデータ形式とは異なる独自の特徴を持っています。

| **特徴** | **時系列データ** | **その他のデータ形式** |
| :---: | :---: | :---: |
| 時間の順序    | 重要              | 重要でない場合が多い                        |
| 時間依存分析  | 必須              | 不要                                         |
| データの相関性| 時間的に関連      | 独立している場合が多い                      |
| 時間スケール  | 分析結果に影響    | 影響しない                                  |
| 保存形式      | <font color="red">タイムスタンプ付き</font>| タイムスタンプ不要                          |

<br>
例：売上データ. （時系列データ）<br>
タイムスタンプあり

```
日付, 売上
2025-01-01, 5000
2025-01-02, 5200
```

例：製品リストデータ. （時系列データではない）<br>
タイムスタンプなし

```
製品ID, 製品名, 価格
001, スマートフォン, 70000
002, ノートパソコン, 120000
```

## <font color="SeaGreen">売上データ 1</font>
以下はKaggleの「Walmart Sales Dataset of 45 stores」データセットで、アメリカの大手小売業者であるWalmartの45店舗における2010年2月5日から2012年10月26日までの1週間ごとの売上データです。<br>
参照）https://www.kaggle.com/datasets/varsharam/walmart-sales-dataset-of-45stores?resource=download

このデータセットについて週ごとの売上推移を可視化してみます。


### 評価ポイント
- 季節性：9月～12月にかけて売上が上昇するパターンが見られ、ホリデーシーズンの影響が考えられます。ブラックフライデーや感謝祭、クリスマス、年末商戦による売上上昇が予想されます。

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# データのロード
df = pd.read_csv('/content/drive/MyDrive/time_series/seminar/files/walmart-sales-dataset-of-45stores.csv')

df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

# 週ごとの売上合計を計算
df['Week'] = df['Date'].dt.to_period('W')
weekly_sales = df.groupby('Week')['Weekly_Sales'].sum()

# 週ごとの売上推移をプロット
plt.figure(figsize=(20, 5))
plt.plot(weekly_sales.index.astype(str), weekly_sales.values, marker='o', linestyle='-')
plt.xlabel("Week")
plt.ylabel("Total Weekly Sales")
plt.title("Weekly Sales Over Time")

# 9月から12月の範囲を抽出
weeks_str = weekly_sales.index.astype(str)
filtered_weeks = [w for w in weeks_str if "-08" in w or "-09" in w or "-10" in w or "-11" in w or "-12" in w]

# 間隔を適切に調整して表示
step = max(1, len(filtered_weeks) // 10)
plt.xticks(ticks=[weeks_str.tolist().index(w) for w in filtered_weeks[::step]],
               labels=filtered_weeks[::step],
               rotation=45, fontsize=10)

# 売上の大きく突出したポイントに日付ラベルを右側に表示
for i, (week, sales) in enumerate(zip(weekly_sales.index.astype(str), weekly_sales.values)):
    if sales > weekly_sales.mean() + 2 * weekly_sales.std():  # 大きく突出している点を検出
        plt.annotate(f'{week}', xy=(i, sales), xytext=(i + 1, sales),
                         arrowprops=dict(facecolor='red', arrowstyle='->'), fontsize=9, color='red', ha='left')


plt.grid()
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime
import warnings
warnings.filterwarnings("ignore")

# データのロード
dateparse = lambda dates: datetime.strptime(dates, '%Y-%m')
data = pd.read_csv('/content/drive/MyDrive/time_series/seminar/files/sales_data.csv',
                   index_col='Date',
                   parse_dates=True,
                   date_parser=dateparse)

# プロット
plt.figure(figsize=(16, 4))
plt.plot(data.index, data.values)

plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=6))  # 6ヶ月ごと
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))  # 年-月形式で表示

plt.xlabel('Date')
plt.ylabel('Service industry sales (million)')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
````

## <font color="SeaGreen">株価データ</font>
yfinance ライブラリを利用して株価データを取得します。yfinance ライブラリは、Yahoo Finance から株価や金融データを簡単に取得できるPythonライブラリです。時系列データのダウンロードや解析に便利で、株価データ、為替データ、暗号通貨データなどを取得するために広く使われています。
2022年1月1日から2024年12月31日までのデータを可視化してみます。

### 評価ポイント
- 明確な周期性はない：季節ごとに株価が上がったり下がったりする決まった周期性は見られません。
- 長期的なトレンドはあるようだが、短期的な動きはランダム：長い目で見ると、上昇・下降のトレンドが**あることもある**が、日ごとの株価の変動は予測が難しく、ランダムに動いているように見えます。

### 参考
**ランダムウォークとは**<br>
株価の動きは「ランダムウォーク」と呼ばれる数学モデルに近い性質を持っています。
ランダムウォークとは、次の位置が確率的に無作為（ランダム）に決定される運動のことです。例えば、コイン投げによって1歩進むか戻るかを決めるモデルが挙げられます。
このようなモデルでは、現在の状態から次の状態への変化が完全にランダムであり、過去の状態に依存しません。この性質は「マルコフ性」と呼ばれ、ランダムウォークはマルコフ連鎖の一種とされています。

ランダムウォークは、物理学や経済学など多くの分野で応用されており、特に株価の変動モデルとしても知られています。株価の値動きは予測が難しく、ランダムウォークに近い挙動を示すと考えられることがあります。

````python
import yfinance as yf
import matplotlib.pyplot as plt

# Apple の株価データを取得
ticker = 'AAPL'
start_date = '2022-01-01'
end_date = '2024-12-31'

# データのダウンロード
data = yf.download(ticker, start=start_date, end=end_date)

# 終値（Close）のプロット
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Close'], label='AAPL Closing Price', color='b')
plt.xlabel('Date')
plt.ylabel('Stock Price (USD)')
plt.title('AAPL Stock Price (2022-2024)')
plt.legend()
plt.grid(True)
plt.show()
````
